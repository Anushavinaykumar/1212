```python
import pandas as pd

data = {
    'Name': ['Rohan', 'Karthik', 'Shymala', 'Unnathi'],
    'Maths': [85, 75, 96, 87],
    'Science': [95, 50, 65, 80],
    'Social': [50, 55, 56, 60]
}

df = pd.DataFrame(data)
print(df)
```

```python
pivot_table = df.pivot_table(index='Name', aggfunc='mean')
print(pivot_table)
```

```python
new_data = {
    'Name': 'NewStudent',
    'Maths': 90,
    'Science': 88,
    'Social': 70
}

df = df.append(new_data, ignore_index=True)
print(df)
```

```python
sorted_df = df[df['Maths'] > 80].sort_values(by='Maths', ascending=False)
print(sorted_df)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.bar(df['Name'], df['Maths'])
plt.xlabel('Student Name')
plt.ylabel('Maths Score')
plt.title('Maths Scores for Students')
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.bar(df['Name'], df['Science'])
plt.xlabel('Student Name')
plt.ylabel('science Score')
plt.title('science Scores for Students')
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.bar(df['Name'], df['Social'])
plt.xlabel('Student Name')
plt.ylabel('Social Score')
plt.title('Social Scores for Students')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.hist(df['Maths'], bins=10, edgecolor='k')
plt.xlabel('Maths Score')
plt.ylabel('Frequency')
plt.title('Histogram of Maths Scores')
plt.show()
```

```python
import matplotlib.pyplot as plt

subject = 'Maths'  # Change to the subject you want to visualize
plt.figure(figsize=(6, 6))
plt.pie(df[subject], labels=df['Name'], autopct='%1.1f%%', startangle=140)
plt.title(f'{subject} Scores')
plt.show()
```

```python
plt.plot(df['Name'], df['Maths'], marker='o', linestyle='-', color='b')
plt.xlabel('Student Name')
plt.ylabel('Maths Score')
plt.title('Maths Scores by Student')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
import pandas as pd
data = {
    'Ename':['Roshan','Amar','ashwini','lohith','Mohan','Pramod'],
    'Type':['Regular','Adhoc','Regular','Adhoc','Contract','regular'],
    'Department':['CS','CS','EC','EC','CS','EC'],
    'Experience':[10,20,5,14,9,8],
    'Salary':[50000,15000,30000,15000,10000,40000]
}
df = pd.DataFrame(data)
print(df)
```

```python
pivot_table_department = df.pivot_table(values='Salary', index='Ename', columns='Department', aggfunc='mean')
print(pivot_table_department)
```

```python
pivot_table_empolyee = df.pivot_table(values='Salary', index='Ename', aggfunc='mean','sum','count')
print(pivot_table_empolyee)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 5))
plt.bar(df['Ename'], df['Salary'])
plt.xlabel('Ename')
plt.ylabel('Salary')
plt.title('Average of Salary')
plt.xticks(rotation=45)
plt.show()
```
