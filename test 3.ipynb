```python
import numpy as np
from sklearn.metrics import mean_absolute_error
actual_value = [2, 3, 4, 8, 4]
predicted_value = [4, 5, 9, 1, 7]
mae = mean_absolute_error(actual_value, predicted_value)
print(f"Mean Absoult_error: {mae}")
```

```python
from sklearn.metrics import r2_score
actual_values = [10, 20, 30, 40, 50]
predicted_values = [12, 18, 28, 38, 52]
r2 = r2_score(actual_values, predicted_values)
print(f"R-squared (R2) Score: {r2}")
```

```python
from sklearn.metrics import mean_squared_error
actual_value = [2, 4, 7, 8]
predicted_value = [4, 1, 6, 9]
sqrt = mean_squared_error(actual_value, predicted_value)
rmse = sqrt
print(f"Root mean squared error:{rmse}")
```

# LeaveOneOut

```python
from sklearn.model_selection import LeaveOneOut
from sklearn.linear_model import LinearRegression
loo = LeaveOneOut()
total_mse = 0
for train_index, test_index in loo.split(X):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    model = LinearRegression()
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    total_mse += mse

average_mse = total_mse / len(X)
print("Average Mean Squared Error (LOOCV):", average_mse)
```
